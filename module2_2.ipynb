merging in data frame

```python
import pandas as pd
#creating DF's
df1 = pd.DataFrame({"A":[1,2,3],"B":[11,21,31]})
df2 = pd.DataFrame({"A":[1,2,3],"C":[10,20,30]})
#perform merge
var1 = pd.merge(df1,df2,on="A")
var1
```

using how


```python
#creating DF's
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})
#perform merge(inner)
var1 = pd.merge(df1,df2,on = "A",how ="inner")
var1
```

```python
#creating DF's
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})
#perform merge(outer)
var1 = pd.merge(df1,df2,on = "A",how ="outer")
var1
```

```python
#creating DF's
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})
#perform merge(left)
var1 = pd.merge(df1,df2,on = "A",how ="left")
var1
```

```python
#creating DF's
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})
#perform merge(right)
var1 = pd.merge(df1,df2,on = "A",how ="right")
var1
```

index

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,4],"B":[10,20,30,40]})
pd.merg(df1,df2,left_index = True,right_index = True)
```

suffix(to change columns name)


```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,3,4],"B":[10,20,30,40]})
pd.merge(df1,df2,left_index = True,right_index = True,suffixes=("_id","_rollno"))
```

concat

```python
#concat in list
sr1 = pd.Series([1,2,3])
sr2 = pd.Series([4,5,6])
pd.concat([sr1,sr2])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2 = pd.DataFrame({"A":[1,2,4,5],"B":[10,20,30,40]})
pd.concat([df1,df2],axis = 1)
```
